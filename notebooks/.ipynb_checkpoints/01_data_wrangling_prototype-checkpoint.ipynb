### Tabelite kokkupanek

Ratingud v2lja

```python
df_ratings = ratings.set_index("placeID")[["rating"]]
df_ratings.head()
```

```python
df_ratings.size
```

K6ik feature-id välja

```python
df_features = geoplaces.join(parking)
df_features.head()
```

```python
df = df_ratings.join(df_features)
```

`df` on lõplik feature-ite tabel. Selles konverteerin yleval tehtud märkmete järgi ümber osad feature-id.

```python
df.head()
```

```python
df_smoking_area = pd.get_dummies(df.smoking_area, prefix="smoking_area")
df_smoking_area.head()
```

```python
df_other_services = pd.get_dummies(df.other_services, prefix="other_services")
df_other_services.head()
```

```python
df_dress_code = pd.get_dummies(df.dress_code, prefix = "dress_code")
df_dress_code.head()
```

```python
df_accesibility = pd.get_dummies(df.accessibility, prefix="access")
df_accesibility.head()
```

```python
df_parking_lot = pd.get_dummies(df.parking_lot, prefix = "parking")
df_parking_lot.head()
```

```python
df_encoded = pd.concat([df[["rating"]],
                        df_smoking_area,
                        df_other_services,
                        df_dress_code,
                        df_accesibility,
                        df_parking_lot
                      ],axis=1)
```

```python
df_encoded
```

### Logistiline regressioon

```python
df_encoded.head()
```

```python
df_encoded.columns
```

```python
Y=df_encoded["rating"].values
```

```python
X=df_encoded.drop("rating").values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
```

```python
logreg = LogisticRegression()
```

```python
logreg = logreg.fit(X, y=Y)
```

```python
X[0]
```

```python
logreg._desc_= "Logistic regression model"
logreg._features_ = list(df_encoded.columns)
```
